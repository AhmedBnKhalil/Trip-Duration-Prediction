## Importing <a class="anchor" id="section_1_2"></a>

```python
import pandas as pd
import geopy.distance as gp
import matplotlib.pyplot as plt
from numpy import mean
from numpy import std
```

# Loading data and preprocssing <a class="anchor" id="chapter2"></a>

```python
df=pd.read_csv('val.csv')
```

```python
df
```

## Finding Nulls <a class="anchor" id="section_2_1"></a>

```python
pd.isnull(df).sum()
```

## Creating new colmns one for origin and another for destination <a class="anchor" id="section_2_2"></a>
### origin = (latitude , longitude) same for destination

```python
origin=pd.DataFrame([df.pickup_latitude,df.pickup_longitude])
dist=pd.DataFrame([df.dropoff_latitude,df.dropoff_longitude])
```

### Transpose the data to reshape it <a class="anchor" id="sub_section_2_2_1"></a>

```python
origin=origin.T
dist=dist.T
```

## creating new list that it calculates the destination from the origin,dist columns by using the geopy libaray <a class="anchor" id="section_2_3"></a>

```python
distance=[]
for i in range(len(origin)):
    
    coords_1=(origin.T[i].pickup_latitude , origin.T[i].pickup_longitude)
    coords_2=(dist.T[i].dropoff_latitude , dist.T[i].dropoff_longitude)
    diss=gp.geodesic(coords_1, coords_2).km
    distance.append(diss)
```

## insert the new column to the dataframe <a class="anchor" id="section_2_4"></a>

```python
df['distance']=distance
df
```

## Droping the useless features <a class="anchor" id="section_2_5"></a>

```python
df=df.drop(columns=["pickup_longitude", "pickup_latitude","dropoff_longitude",'dropoff_latitude','id','vendor_id','store_and_fwd_flag'])
```

```python
df['pickup_datetime']=pd.to_datetime(df['pickup_datetime'])
df['dropoff_datetime']=pd.to_datetime(df['dropoff_datetime'])
#df=df[['pickup_datetime', 'dropoff_datetime', 'passenger_count','distance','trip_duration']]
```

## Converting the trip duration to minutes by divide the seaconds by 60 <a class="anchor" id="section_2_6"></a>

```python
df.trip_duration=df.trip_duration/60
```

# Understanding the data <a class="anchor" id="section_3"></a>

```python
df.describe(include='all',datetime_is_numeric=True)
```

```python
df.sort_values(by=['trip_duration'],ascending=False)
```

## Finding the outliers <a class="anchor" id="section_3_1"></a>

```python
# calculate summary statistics
data_mean, data_std = mean(df['trip_duration']), std(df['trip_duration'])
# identify outliers
cut_off = data_std * 3
lower, upper = 0, data_mean + cut_off
lower,upper
...
# identify outliers
outliers = [x for x in df['trip_duration'] if x < 0 or x > upper]
outliers
```

```python
print('Identified outliers: %d' % len(outliers))
...
# remove outliers
outliers_removed = [x for x in df['trip_duration'] if x > lower and x < upper]
print('Non-outlier observations: %d' % len(outliers_removed))
```

## Removing the outliers <a class="anchor" id="section_3_2"></a>

```python
for idx,i in enumerate(df['trip_duration']):
    if i >upper:
        df=df.drop([idx])
df.sort_values(by=['trip_duration'],ascending=False)
```
